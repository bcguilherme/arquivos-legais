# PLotting

```python
ax = df["yearintroduced"].value_counts()\
.head(10)\
.plot(kind="bar",title ="Top Year Coaster")
ax.set_xlabel=("Year Introduced")
ax.set_ylabel=("Count")
```

```python
ax=df["speedmph"].plot(kind="hist", bins=20)
ax.set_title("Speed")
ax.set_xlabel("Speed")
ax.set_ylabel("count")
```

```python
ax=df["speedmph"].plot(kind="kde")
ax.set_title("Speed")
ax.set_xlabel("Speed")
ax.set_ylabel("count")
```
