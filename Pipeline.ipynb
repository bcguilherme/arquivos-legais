```python
from sklearn import set_config
from sklearn.pipeline import Pipeline, make_pipeline
from sklearn.compose import ColumnTransformer, make_column_transformer
from sklearn.preprocessing import StandardScaler, PowerTransformer, OrdinalEncoder, OneHotEncoder
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
set_config(display="diagram")
```

```python
df=pd.read_csv("/Users/chenyenpin/Downloads/titanic.csv", index_col="PassengerId")
df.head()

```

```python
df["personal_tiltes"]=df.Name.str.extract(".*, (.+?)\.", expand=False)
df.head()

```

```python
df=df.drop(columns=["Name", "Ticket","Cabin"], axis=1).dropna()
df.head()
```

```python
target=df["Survived"]
df_features=df.drop(columns="Survived")
df_features.head()
```

```python
#selecionando features numericas e categoricas

numerical_features=df_features.select_dtypes("number").columns
categorical_features=df_features.select_dtypes(exclude="number").columns
```

```python
numerical_features,categorical_features
```

```python
#aplicando column trasnformer
preprossessin_pipeline=make_column_transformer([StandardScaler(), numerical_features],
                        [OneHotEncoder(), categorical_features])
```

```python
preprossessin_pipeline.fit(df_features)
```

```python
pd.DataFrame(
preprossessin_pipeline.transform(df_features).toarray(),
columns=preprossessin_pipeline.get_feature_names_out())
```

```python
#multiplos preporcessamentos no column transformer
#aplicando columns transformer
preprossessin_pipeline=make_column_transformer([make_pipeline(PowerTransformer(),StandardScaler()), numerical_features],
                        [OneHotEncoder(), categorical_features])
```

```python
preprossessin_pipeline.fit(df_features)
```

```python
pd.DataFrame(
data=preprossessin_pipeline.transform(df_features).toarray(),
columns=preprossessin_pipeline.get_feature_names_out())
```
