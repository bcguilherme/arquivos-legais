```python
gaucho
plt.figure(figsize=(8,5))
plt.subplot(1,2,1)
sns.histplot(data=gaucho.PIB, kde=True)
plt.subplot(1,2,2)
sns.histplot(data=gaucho.VALOREMPENHO, kde= True)

```

```python
#ordernado por PIB iloc[:,1:3] usando iloc, todas as linhasm , '[,1:3], selecionando apenas as colunas
# 1 a 3 sendo q coluna 3 nao entra na selecao
ordenado =gaucho.sort_values("PIB").head(10)
#agrupado
ordenado = ordenado.iloc[:,1:3]
ordenado
ordenado.plot.bar(x="MUNICIPIO", y="PIB", color="blue");


```

```python
#ordenando o valor empenho, e depois usando iloc[:,[1:3]], selecionando todas as linhas,
#mas colocando a posicao [:,[1:3]], selecionamos apenas a coluna 1 e 3 da tabela
ordenado_empeenho=gaucho.sort_values("VALOREMPENHO").head(10)
#ordenado_empeenho
ordenado_empeenho=ordenado_empeenho.iloc[:,[1,3]]
ordenado_empeenho.plot.bar(x="MUNICIPIO", y="VALOREMPENHO", color="blue")
```
